# Bài học 4: 4 - Vòng lặp for và while

## Mục tiêu
- Hiểu khái niệm cơ bản
- Viết được ví dụ minh hoạ
- Làm bài tập thực hành

```markdown
### Nội dung bài học

1. Khái niệm vòng lặp trong lập trình.
2. Cú pháp và cách sử dụng vòng lặp for.
3. Cú pháp và cách sử dụng vòng lặp while.
4. So sánh vòng lặp for và while.
5. Ứng dụng vòng lặp for để lặp qua danh sách.
6. Ứng dụng vòng lặp while để lặp đến khi điều kiện sai.
7. Sử dụng break và continue trong vòng lặp.
8. Lồng ghép vòng lặp (vòng lặp trong vòng lặp).
9. Một số lỗi thường gặp khi sử dụng vòng lặp.
10. Bài tập thực hành về vòng lặp for và while.
```

````python
# 1. Khái niệm vòng lặp trong lập trình
print("1. Khái niệm vòng lặp trong lập trình:")
print("Vòng lặp cho phép thực hiện lặp đi lặp lại một khối lệnh cho đến khi điều kiện dừng được đáp ứng.\n")

# 2. Vòng lặp for
print("2. Vòng lặp for - In các số từ 0 đến 4:")
for i in range(5):
    print(i)
print()

# 3. Vòng lặp while
print("3. Vòng lặp while - In các số từ 0 đến 4:")
i = 0
while i < 5:
    print(i)
    i += 1
print()

# 4. So sánh vòng lặp for và while
print("4. So sánh vòng lặp for và while:")
print("- for: Dùng khi biết trước số lần lặp hoặc lặp qua dãy.")
print("- while: Dùng khi chỉ biết điều kiện dừng, không biết trước số lần lặp.\n")

# 5. Ứng dụng vòng lặp for để lặp qua danh sách
print("5. Ứng dụng vòng lặp for để lặp qua danh sách:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
print()

# 6. Ứng dụng vòng lặp while để lặp đến khi điều kiện sai
print("6. Ứng dụng vòng lặp while để lặp đến khi điều kiện sai:")
count = 0
while count < 3:
    print("Lần lặp:", count)
    count += 1
print()

# 7. Sử dụng break và continue trong vòng lặp
print("7. Sử dụng break trong vòng lặp for (dừng khi i == 3):")
for i in range(5):
    if i == 3:
        break
    print(i)
print()

print("Sử dụng continue trong vòng lặp for (bỏ qua khi i == 2):")
for i in range(5):
    if i == 2:
        continue
    print(i)
print()

# 8. Lồng ghép vòng lặp (vòng lặp trong vòng lặp)
print("8. Lồng ghép vòng lặp (for trong for):")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
print()

# 9. Một số lỗi thường gặp khi sử dụng vòng lặp
print("9. Một số lỗi thường gặp khi sử dụng vòng lặp:")
print("- Quên cập nhật biến điều kiện trong while → vòng lặp vô hạn.")
print("- Sai phạm vi lặp trong for.")
print("- Sử dụng sai break/continue.\n")

# 10. Bài tập thực hành (gợi ý)
print("10. Bài tập thực hành:")
print("- In ra các số chẵn từ 1 đến 10 bằng vòng lặp for.")
print("- Tính tổng các số từ 1 đến n bằng vòng lặp while.")
print("- Duyệt qua một danh sách và in ra các phần tử có độ dài lớn hơn 3 ký tự.")
````
