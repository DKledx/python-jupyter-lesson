```python
# 1. Hiểu khái niệm về câu lệnh điều kiện trong Python
print("1. Câu lệnh điều kiện cho phép chương trình quyết định thực hiện một đoạn mã dựa trên điều kiện đúng/sai.")

# 2. Phân biệt giữa if, elif và else
print("\n2. if kiểm tra điều kiện đầu tiên, elif kiểm tra điều kiện tiếp theo nếu if sai, else thực hiện khi tất cả điều kiện trên đều sai.")
x = 5
if x > 10:
    print("x lớn hơn 10")
elif x == 5:
    print("x bằng 5")
else:
    print("x nhỏ hơn hoặc bằng 10 và khác 5")

# 3. Cú pháp cơ bản của câu lệnh if
print("\n3. Cú pháp cơ bản:")
if x == 5:
    print("x là 5")

# 4. Sử dụng toán tử so sánh trong điều kiện
print("\n4. Toán tử so sánh: ==, !=, >, <, >=, <=")
y = 7
if y != 5:
    print("y khác 5")

# 5. Kết hợp nhiều điều kiện với toán tử logic (and, or, not)
print("\n5. Kết hợp điều kiện với and, or, not:")
a, b = 3, 8
if a < 5 and b > 5:
    print("a nhỏ hơn 5 và b lớn hơn 5")

# 6. Ứng dụng câu lệnh điều kiện trong giải quyết bài toán thực tế
print("\n6. Ứng dụng thực tế: Kiểm tra số chẵn/lẻ")
n = 4
if n % 2 == 0:
    print(f"{n} là số chẵn")
else:
    print(f"{n} là số lẻ")

# 7. Lồng ghép các câu lệnh điều kiện (nested if)
print("\n7. Lồng ghép if:")
score = 85
if score >= 50:
    if score >= 80:
        print("Đạt loại Giỏi")
    else:
        print("Đạt loại Khá")
else:
    print("Chưa đạt")

# 8. Thực hành viết các ví dụ minh hoạ với if, elif, else
print("\n8. Ví dụ: Xếp loại học lực")
diem = 7.5
if diem >= 8:
    print("Giỏi")
elif diem >= 6.5:
    print("Khá")
elif diem >= 5:
    print("Trung bình")
else:
    print("Yếu")

# 9. Phân tích lỗi thường gặp khi sử dụng câu lệnh điều kiện
print("\n9. Lỗi thường gặp: Thiếu dấu hai chấm, thụt lề sai")
# Ví dụ lỗi:
# if x == 5
# print("x là 5")  # Thiếu dấu hai chấm và thụt lề

# 10. Bài tập vận dụng và thảo luận kết quả
print("\n10. Bài tập: Nhập một số và kiểm tra số đó là dương, âm hay bằng 0")
num = -3
if num > 0:
    print("Số dương")
elif num < 0:
    print("Số âm")
else:
    print("Số 0")
```
